```python
import matplotlib.pyplot as plt
import numpy as np
import math
from math import comb

def Binomial(n,p):
    x = np.arange(0,n,1)   # ASSUMING X VALUES FROM 0 TO N AND WE CHOOSE 1 VALUE FROM IT
    f = lambda x,n,p: comb(n,x) * (p**x) * (1-p)**(n-x)    # DEFINING FUNCTION 
    y = [f(i,n,p) for i in x]      # COMPUTING F(X) VALUES
    plt.title('Binomial Distribution')
    plt.xlabel('X values from 0-15')
    plt.ylabel('f(x) Values')
    plt.bar(x,y)
    plt.show()
    
Binomial(15,0.5)
```

# Poisson Distribution : 
    

$f = lambda λ,x: math.exp(-1*λ) * (λ**x/math.factorial(x))$

```python
def Poisson(λ):
    x = np.arange(0,n,1)
    f = lambda λ,x: math.exp(-1*λ) * (λ**x/math.factorial(x))
    y = [f(λ,i) for i in x]
    plt.title('Poisson Distribution')
    plt.ylabel('f(x) Values')
    plt.bar(x,y)
    plt.show()
Poisson(1)
```

# Plotting Exponential Dsitribution : 
   

 $f = lambda  λ,x: λ*(math.exp(-1*λ*x))$

```python
def Exponential(λ):
    x = np.arange(0,n,1)
    f = lambda λ,x: λ*(math.exp(-1*λ*x))
    y = [f(λ,i) for i in x]
    plt.title('Exponential Distribution')
    plt.xlabel('X values from 0-15')
    plt.ylabel('f(x) Values')
    plt.bar(x,y)
    plt.show()
Exponential(100)
```

Normal Distribution
   $ f = lambda x,μ,σ: (1/(sqrt(2*math.pi*σ)))*math.exp(-1/2(x-μ)**2) $

```python
def Normal(m,v):
    x = np.arange(0,n,1)
    a = (1/(math.sqrt(2*(math.pi)*v)))
    f = lambda x,m,v: a*math.exp((-1/2)*(((x-m)**2)/v))
    y = [f(i,m,v) for i in x]
    plt.title('Normal Distribution')
    plt.ylabel('f(x) Values')
    plt.plot(x,y)
    plt.show()
Normal(6,25)
```

```python
def Gamma(a,l,n):
    z = np.arange(-n,n,1)
    p = (l**a)/(math.factorial(a-1)) 
    f = lambda x,l,a: p * (x**(a-1)) * math.exp(-1*l*x)
    y = [f(i,a,l) for i in z]
    plt.title('Gamma Distribution')
    plt.plot(z,y)
    plt.show()
Gamma(3,7,100)
```
