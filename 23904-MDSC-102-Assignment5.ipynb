```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
Names = ["class", "Alcohol",
 	 "Malic acid",
 	 "Ash",
	 "Alcalinity of ash",  
 	 "Magnesium",
	 "Total phenols",
 	 "Flavanoids",
 	 "Nonflavanoid phenols",
 	 "Proanthocyanins",
	 "Color intensity",
 	 "Hue",
 	 "OD280/OD315 of diluted wines",
 	 "Proline"   ]
df = pd.read_csv("wine.data",names = Names)
df
```

```python
sns.boxplot(data=df,x='class',y='Alcohol',hue='class',linewidth=1)
plt.show()
```
